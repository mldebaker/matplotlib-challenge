```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse = pd.read_csv('data/mouse_drug_data.csv')
trial = pd.read_csv('data/clinicaltrial_data.csv')

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse.head()
trial.head()

# Combine the data into a single dataset
df = pd.merge(trial, mouse)

# Display the data table for preview
df.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
df2 = df.drop(columns=['Mouse ID', 'Metastatic Sites'])

# Convert to DataFrame
df3 = df2.groupby(['Drug', 'Timepoint']).mean()

# Preview DataFrame
df3.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame
standard = df2.groupby(['Drug', 'Timepoint']).sem()
# Preview DataFrame
standard.head()
```
