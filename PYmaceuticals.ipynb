   # 45 Day Tumor Treatment On Mice 

#### Observations of Mice Study

   The first trend that stands out clearly is that the drug Capomulin was the only succesful drug at reducing the size of the tumor in the mice out of the drugs we decided to analyze. Capomulin was also very succesful at reducing metastatic cells from spreading along with the drug Infubinol. Another trend that i noticed while analyzing the data was that the drug Ketapril actually increased the size of the tumor when compared to the placebo but still had a reduction in metastatic spread of the cells. The drug Infubinol seemed to increase the death rate of the mice greatly after 30 days of treatment when compared to the placebo. 

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse = pd.read_csv('data/mouse_drug_data.csv')
trial = pd.read_csv('data/clinicaltrial_data.csv')

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse.head()
trial.head()

# Combine the data into a single dataset
df = pd.merge(trial, mouse)

# Display the data table for preview
df.head()
```

# Tumor Response to Drug Treatment

```python
# Mean Tumor Volume Data Grouped by Drug and Timepoint 
df2 = df.drop(columns=['Mouse ID', 'Metastatic Sites'])
df3 = df2.groupby(['Drug', 'Timepoint']).mean()

# Preview DataFrame
df3.head()
```

```python
# Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard = df2.groupby(['Drug', 'Timepoint']).sem()

# Preview DataFrame
standard2 = standard.reset_index()
standard3 = standard2.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')
standard3
```

```python
# Data Munging to Re-Format the Data Frames
avgTumor = df3.reset_index()
avgTumor1 = avgTumor.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
avgTumor1.head()
```

```python
# Generate the Plot (with Error Bars)
drugs = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]
for Drug,marker,colors in drugs:
    plt.errorbar(avgTumor1.index,avgTumor1[Drug],yerr = standard3[Drug],fmt = '--', marker = 'o')
plt.ylim(25, 80)
plt.legend(loc='best')
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid()

# Save the Figure
plt.savefig('Tumor_Response_to_Treatment.png')

# Display Figure
plt.show()
```

# Metastatic Response to Treatment

```python
# Mean Met. Site Data Grouped by Drug and Timepoint 
met = df.drop(columns=['Mouse ID', 'Tumor Volume (mm3)'])
met1 = met.groupby(['Drug', 'Timepoint']).mean()

# Preview DataFrame
met1.head()
```

```python
# Data Munging to Re-Format the Data Frames
met2 = met1.unstack(level = 'Drug')
met2.columns = met2.columns.get_level_values(1)

# Preview that Reformatting worked
met2
```

```python
# Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_met = df.drop(columns=['Mouse ID', 'Tumor Volume (mm3)'])
sem_met1 = sem_met.groupby(['Drug', 'Timepoint']).sem()

# Preview DataFrame
sem_met1.head()
```

```python
# Data Munging to Re-Format the Data Frames
sem_met2 = sem_met1.unstack(level = 'Drug')
sem_met2.columns = sem_met2.columns.get_level_values(1)

# Preview that Reformatting worked
sem_met2
```

```python
# Generate the Plot (with Error Bars)
drugs = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]
for Drug,marker,colors in drugs:
    plt.errorbar(met2.index,met2[Drug],yerr = sem_met2[Drug],fmt = '--', marker = 'o')

plt.legend(loc='best')
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Metastatic Sites')
plt.grid()

# Save the Figure
plt.savefig('Metastatic Sites.png')

# Display Figure
plt.show()
```

# Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse = df.drop(columns=['Metastatic Sites', 'Tumor Volume (mm3)'])
mouse1 = mouse.groupby(['Drug', 'Timepoint']).count()

# Preview DataFrame
mouse1.head()
```

```python
# Data Munging to Re-Format the Data Frames
mouse2 = mouse1.unstack(level = 'Drug')
mouse2.columns = mouse2.columns.get_level_values(1)

# Preview the Data Frame
mouse2
```

```python
# Generate the Plot (with Error Bars)
drugs = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]
for Drug,marker,colors in drugs:
    plt.errorbar(mouse2.index,mouse2[Drug],fmt = '--', marker = 'o')

plt.legend(loc='best')
plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')
plt.grid()

# Save the Figure
plt.savefig('Survival_During_Treatment.png')

# Display Figure
plt.show()
```

# Summary Bar Graph (45 Day Treatment)

```python
# Calculate the percent changes for each drug
avgTumor2 = avgTumor1.loc[:,['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
drugp = (avgTumor2.loc[45]-avgTumor2.loc[0])/avgTumor2.loc[0]*100

# Display the data to confirm
drugp
```

```python
# Create Bar Graph
percent2 = drugp.index
plt.grid()
plt.rc('axes', axisbelow=True)
plt.xticks(np.arange(len(percent2)), percent2)
plt.bar(np.arange(4),drugp,color="grrr")
counter = 0
for value in drugp:    # Loop For Percent                                                   
    if value < 0:                                                          
        y_point = -7
    else:
        y_point = 7
    plt.text(counter, y_point, str(int(value)) + '%', ha = 'center', color = 'black')
    counter += 1                                                  
plt.axhline(y=0, color = 'black') 
plt.ylabel("% Tumor Volume Change", fontsize=13)
plt.xlabel("Test Drugs", fontsize=13)
plt.title("Tumor Change Over 45 Day Treatment", fontsize=15)

# Save the Figure
plt.savefig('Bar_Graph_45_Day_Treatment.png')

# Display Figure
plt.show()
```
